Supervised learning - learns from labeled examples, like a student studying with answers provided. Needs prelabeld data(think flashcards)

Self - supervised learning - trains on unlabeled data by creating its own supervisory signals.Two common approaches: reconstruction, constrasive learning

Reinforcement learning - learns through trial and error , like an animal receiving rewards for desired actions. Interacts with an environment and gets feedback (rewards or penalties) to improve.

Multi layer perceptron - This can learn nonlinear functions provided each perception has a differentiable nonlinearity

Generative models - If we study the probabilty distribution function of data, we can generate similar models

Generative models learn the underlying distribution of data. Once learned , they can generate new data samples that resemble the training data.

```python
import cv2 as cv
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('D:\semester 5\EN3160\images\emma.jpg' , cv.IMREAD_GRAYSCALE)
assert im is not None

im2 = im + 100
#When you use im + 100, you're performing element-wise addition using NumPy. If im is an 8-bit image,
#the addition might result in values that exceed the 8-bit range (255). 
#In such cases, the values will "wrap around" due to modulo behavior. 
#For example, adding 100 to a pixel value of 200 would result in 44 (300 % 256 = 44), not 255. 
#This is known as modulo arithmetic.

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
ax[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB), cmap=cm.gray, vmin=0, vmax=255)
ax[0].axis('off')
ax[1].imshow(cv.cvtColor(im2, cv.COLOR_BGR2RGB),cmap=cm.gray, vmin=0, vmax=255)
ax[1].axis('off')
plt.show()
```

```python
import cv2 as cv
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('D:\semester 5\EN3160\images\emma.jpg' , cv.IMREAD_GRAYSCALE)
assert im is not None

im2 = cv.add(im,100)
#The cv.add() function ensures that the result is clipped to the valid range for image data types. 
#For example, if im is an 8-bit unsigned integer image (where pixel values range from 0 to 255)
#adding 100 to a pixel with a value of 200 would result in 255 (not 300) because the maximum value for a pixel is 255.
#This is known as saturation arithmetic.

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
ax[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB), cmap=cm.gray, vmin=0, vmax=255)
ax[0].axis('off')
ax[1].imshow(cv.cvtColor(im2, cv.COLOR_BGR2RGB),cmap=cm.gray, vmin=0, vmax=255)
ax[1].axis('off')
plt.show()
```
